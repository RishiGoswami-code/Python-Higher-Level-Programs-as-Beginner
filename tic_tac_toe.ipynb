The logic which are used in Tic-Tac-Toe Code
1.Making A Nested Matrix For Board
The board is created as a 3 rows using nested lists. Each cell is initially empty (represented by a space '
').
board = [[' ' for _ in range(3)] for _ in range(3)] {This format of nested loop was taught by prof. Pintu sir}
2. Printing the Board
Defined the print_board function to show the matrix in the tabular form not in the horizontal form.
Logic: Iterate through each row of the board. And printed each row.
3. Player Move Selection
Logic:
- this function takes two input one is pisotion which user want to fill and the coin or symbol which is
specified for each players.
- If the position matches the player's choice, update the cell with their mark ('X' or 'O'). {board[i][j]=coin}
4. Players function
Logic:
Check if the decision of the player exists in the list of available matrix or not. If available, update the
board and remove the position from the available list.
If invalid, ask again to play the player.
5. Checking Win Conditions
The win_lose function determines if a player has won the game. It checks rows, columns, and diagonals
to see if all cells in any line contain the same mark.
Logic:
- Check each row to see if all elements are the same as the player's coin or symbol.
- Check each column similarly.
- Check the two diagonals (main and anti-diagonal) for the same condition.
6. Alternating Turns
The play_game function alternates turns between Player 1 and Player 2.
Logic:
- Use a loop to alternate between Player 1 and Player 2 for up to 9 moves.
- After each move, check for a winning condition using win_lose.
- If a player wins, end the game and announce the winner.

```python
board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board(board):                           # Defining the matrix as board for the game
    for row in board:
        print(row)

print_board(board)

pos = []
for i in range(1, 10):                           # Defining the positions of the board
    pos.append(i)

def choose(pos):
    print(f"{pos} positions are empty")          # Showing the empty spaces

def selection(position, coin):                   # This function is made for giving the input like X or O
    c = 0
    for i in range(len(board)):
        for j in range(len(board)):
            c += 1
            if c == position:
                board[i][j] = coin

def win_lose(board, coin):                       # This function checks if a player has won
    for row in board:                            # Check rows
        if row[0] == row[1] == row[2] == coin:
            return True
    for col in range(3):                         # Check columns
        if board[0][col] == board[1][col] == board[2][col] == coin:
            return True
    
    if board[0][0] == board[1][1] == board[2][2] == coin:  # Check main diagonal
        return True

    if board[0][2] == board[1][1] == board[2][0] == coin:  # Check anti-diagonal
        return True

    return False

def player1():                                   # Defining the first player to play
    decision = int(input("Enter the position which you want to select:  "))
    if decision in pos:
        selection(decision, "X")
        pos.remove(decision)
        return board
    else:
        print("Occupied! Please enter a valid place")
        player1()
        
def player2():                                   # Defining the second player to play
    decision = int(input("Enter the position which you want to select:   "))
    if decision in pos:
        selection(decision, "O")
        pos.remove(decision)
        return board
    else:
        print("Occupied! Please enter a valid place")
        player2()
        
if __name__ == "__main__":

    def play_game():
        for i in range(1, 10):
            if i % 2 == 0:
                player1()
                print_board(board)
                if win_lose(board, "X"):
                    print_board(board)
                    print("Player 2 (X) wins!")
                    return
            else:
                player2()
                print_board(board)
                if win_lose(board, "O"):
                    print_board(board)
                    print("Player 2 (O) wins!")
                    return
            choose(pos)  
        print("It's a draw!")
    
    play_game()
```
